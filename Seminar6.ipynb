## Distributia normala

````python
from scipy.stats import norm
````

````python
Z=norm()
````

````python
Z.mean()
````

````python
Z.std()
````

````python
x=np.linspace(-4,4,100)
plt.plot(x, Z.pdf(x))
#plt.plot(x, Z.pdf(x), 'o') # 'o' - marcar
````

### Functia de repartitie - reprezentare

````python
x=np.linspace(-4,4,100)
plt.plot(x, Z.cdf(x))
````

````python
Z.cdf(0)
````

$P(Z < 1.3)=?$

````python
Z.cdf(1.3)
````

````python
x=np.linspace(-4,4,100)
plt.plot(x, Z.pdf(x))
plt.fill_between(x, Z.pdf(x), where = x<1.3) # colorarea curbei
plt.title(f'$P(Z<1.3)=${Z.cdf(1.3)}')
````

$P(Z \geq 1.3)=?$

````python
1-Z.cdf(1.3)
````

````python
x=np.linspace(-4,4,100)
plt.plot(x, Z.pdf(x))
plt.fill_between(x, Z.pdf(x), where = x>1.3)
plt.title(f'$P(Z>1.3)=${1-Z.cdf(1.3)}')
````

$P(0.2 \le Z \le 0.8)=?$

````python
Z.cdf(0.8)-Z.cdf(0.2)
````

````python
x=np.linspace(-4,4,100)
plt.plot(x, Z.pdf(x))
plt.fill_between(x, Z.pdf(x), where = np.logical_and((x>0.2) , (x<0.8))) # "logical_and" returneaza valorile lui x pentru care ambele conditii sunt indeplinite
plt.title(f'$P(0.2<Z<0.8)=${Z.cdf(0.8)-Z.cdf(0.2)}')
````

Care este probabilitatea ca Z sa ia valori in afara intervalului [0.2, 0.8]?

````python
Z.cdf(0.2) + (1-Z.cdf(0.8))
````

````python
x=np.linspace(-4,4,100)
plt.plot(x, Z.pdf(x))
plt.fill_between(x, Z.pdf(x), where = np.logical_or((x<0.2) , (x>0.8)))
plt.title(f'$P(Z<0.2 or Z>0.8)=${1-(Z.cdf(0.8)-Z.cdf(0.2))}')
````
